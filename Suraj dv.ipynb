```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv("Grocery_data.csv")
data.dropna()
df=data.head(20)
```

```python
# 1. Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Product Name', y='Original Price (Rs.)', data=df, color='skyblue', label='Original Price')
sns.barplot(x='Product Name', y='Discounted Price (Rs.)', data=df, color='red', label='Discounted Price')
plt.xticks(rotation=90)
plt.title('Original Price vs. Discounted Price')
plt.xlabel('Product Name')
plt.ylabel('Price (Rs.)')
plt.legend()
plt.show()
```

```python
# 2. Histogram
df=data.head(14)
plt.figure(figsize=(8, 5))
sns.histplot(df['Discount'].str.replace('% OFF', '').astype(float), bins=10, kde=True)
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Pie Chart
df=data.head(650)
plt.figure(figsize=(8, 8))
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Products by Category')
plt.ylabel('')
plt.show()
```

```python
# 4. Box Plot
df=data.head(100)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['Original Price (Rs.)', 'Discounted Price (Rs.)']])
plt.title('Box Plot of Original Price and Discounted Price')
plt.ylabel('Price (Rs.)')
plt.show()
```

```python
# 5. Scatter Plot
df=data.head(60)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Original Price (Rs.)', data=df)
plt.xticks(rotation=45, ha='right') 
plt.title('Quantity vs. Original Price')
plt.xlabel('Quantity')
plt.ylabel('Original Price (Rs.)')
plt.show()
```

```python
# 6. Violin Plot
df=data.head(500)
plt.figure(figsize=(20, 6))
sns.violinplot(x='Category', y='Discounted Price (Rs.)', data=df)
plt.title('Distribution of Discounted Prices Across Categories')
plt.xlabel('Category')
plt.ylabel('Discounted Price (Rs.)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Heatmap
df=data
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Quantity', 'Original Price (Rs.)', 'Discounted Price (Rs.)']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
